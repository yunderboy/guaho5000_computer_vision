# BGR to RGB

```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np

import time as t
print("Running OpenCV Version : %s " % cv2.__version__)
```

```python
def BGR_to_RGB(img, info=False):
    t0 = t.time()
    cv_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    t1 = t.time()
    dt_cv = t1-t0
    
    if info:
        print("Conversion took %0.5f seconds" % dt_cv)
    
    return cv_rgb
```

```python
plt.imshow(BGR_to_RGB(cv2.imread('images/cat.jpeg'), info=True))
plt.show()
```
